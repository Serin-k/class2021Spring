## **선형대수2**

#### **Vector Space**
R^n
- R^1: 1차원
- R^2: 2차원
- R^n: n차원. 숫자가 n개가 있을 때


#### **Column Space**

두개의 벡터 (Col1 & Col2)는 한 선상에 
- 놓여잇지 않다 -> independent
- 놓여있다 -> dependent

- 모든 column vector즉 linear combination이 한 선상에 있다면?
  - linear combination: c\*v + d*w
  - v, w는 벡터
- independent한 두 점의 벡터를 무한대로 보내 어떤수를 곱하고 둘을 더하고 하면 공간속의 모든 점을 다 만들 수 있다
  - 공간속의 vector는 linear combination으로 채울 수 있다
- 근데 한 선상에 있는 dependent할 때에는 두개를 더한다고 치더라도 공간을 다 채울 수 없다. 아무리 해도 선상에서 벗어날 수 없다
